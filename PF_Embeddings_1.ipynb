```python
from langchain.vectorstores import Chroma
```

```python
from langchain.embeddings import SentenceTransformerEmbeddings
```

```python
directory = "/content/drive/MyDrive/PF_PDFs/Embeddings/PF_Embeddings"
embeddings = SentenceTransformerEmbeddings(model_name ='sentence-transformers/all-MiniLM-L6-v2')

vectordb = Chroma(persist_directory=directory, embedding_function=embeddings)

query = input("Type your search: ")

result_docs = vectordb.similarity_search_with_score(query)

for result in result_docs:
    source = result[0].metadata
    print(f'Source: {source}')
    content = result[0].page_content
    print(f'Content: {content}')
    print('')
```

```python
def chroma_search():
    embeddings = SentenceTransformerEmbeddings(model_name ="sentence-transformers/all-MiniLM-L6-v2")
    vectorDB = Chroma(persist_directory=directory, embedding_function=embeddings)
    return vectorDB

vectorDB = chroma_search()

result_docs = vectorDB.similarity_search_with_score(query,k=1)
result_docs[0][0].page_content
score = result_docs[0][1]
source = result_docs[0][0].metadata
```
